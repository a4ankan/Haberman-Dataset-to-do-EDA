What is the data set all about??
---------------------------------------------
It is a dataset which tells us about the no of paitents who survived a breast cancer surgery that was conducted b/w 1958-1970 at the University of Chicago's billlings hospital.

Attribute Information:
-------------------------------

1. Age of patient at time of operation (numerical)
2. Patient's year of operation (year - 1900, numerical)
3. Number of positive axillary nodes detected (numerical)
4. Survival status (class attribute)
(1) = Survived - the patient survived 5 years or longer
(2) = Not Survived - the patient died within 5 year

```python
survived = f.loc[f['status'] == 1]
dead = f.loc[f['status'] == 2]


counts, bin_edges = np.histogram(survived['pos_axillary_nodes'], bins=20, 
                                 density = True,)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r')
plt.plot(bin_edges[1:], cdf)
plt.xlabel('pos_axillary_nodes')
plt.legend(['pdf_survived', 'cdf_survived'])
plt.show()
```

```python
counts, bin_edges = np.histogram(dead['pos_axillary_nodes'], bins=20, 
                                 density = True,)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'g')
plt.plot(bin_edges[1:], cdf,)
plt.legend(['pdf_dead', 'cdf_dead'])
plt.xlabel('pos_axillary_nodes')
plt.show()
```

Observations
